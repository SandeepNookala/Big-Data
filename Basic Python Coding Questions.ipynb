```
#math operations union,intersection,difference,symmetric difference

s2 = set(i for i in range(10) if i%2==0)
s3 = set(i for i in range(10))

print(s2,type(s2))
print(s3,type(s3))

U = s2.union(s3)
print(U)


I = s2.intersection(s3)
print(I)

d = s3.difference(s2)
print(d)


sd = s2.symmetric_difference(s3)
print(sd)
```

```
#write a function to find factorial of given number using function and while loop?

def factorial(n):
    num=1
    while n>=1:
        num = num*n
        n=n-1
    return num
    
factorial(6)
```

```
#write a function to find factorial of given number using function and for loop?

def fact(n):
    num =1
    for i in range(1,n+1):
        num = num*i
    print('factorial',num)
fact(6)
```

```
#filter even and odd values from tuple of 100 range 


tu = {i for i in range(100)}

# Even values
def even(x):
    if x%2 ==0:
        return True
    else :
        return False

tu_ev = tuple(filter(even,tu))

print(tu_ev)

#odd values

def odd(x):
    if x%2 !=0:
        return True
    else:
        return False
        
tu_od = tuple(filter(odd,tu))

print(tu_od)
```

```
#filter even and odd values from tuple of 10 range using lambda

even = tuple(filter(lambda x: x%2 ==0,tu))  #filter values from given sequence based on condition
print(even)

odd = tuple(filter(lambda x: x%2 !=0,tu))
print(odd)
```

```
#print square of given tuple using lambda and map

tu_sq = tuple(map(lambda x: x*x ,tu_ev))  #apply some functionality on every element in sequence generates new element

print(tu_sq)
```

```
#multiple two list using lambda and map

tu_mul = tuple(map(lambda x,y: x*y,tu_ev,tu_od))

print(tu_mul)
```

```
# Sum ,multipile all elements in list using reduce function

from functools import *

tu_muli = reduce(lambda x,y:x*y,tu_od)   #reduce to single output
print(tu_muli)

tu_sum = reduce(lambda x,y:x+y ,tu_od)

print(tu_sum)
```
