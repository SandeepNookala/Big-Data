```
#write a function to find factorial of given number using function and while loop?

def factorial(n):
    num=1
    while n>=1:
        num=num*n
        n=n-1
    return num
factorial(5)      
    
```

```
#write a function to find factorial of given number using function and for loop?
def factorial(n):
    num=1
    for i in range(1,n+1):
        num=num*i
    print("factorial of",n, "is :",num)
factorial(5)
```

```
# filter even values in given list in range 10

def even(x):
    if x%2==0:
        return True
    else:
        return False
    

L=[i for i in range(1,10)]
L1=list(filter(even,L))
print(L1)
```

```
#filter even values from list of 100 range using lambda

L2=list(filter(lambda x:x%2 == 0,L))
print(L2)
```

```
#filter odd values from tuple of 10 range 

def odd(x):
    if x%2 !=0:
        return True
    else:
        return False

T=tuple(i for i in range(1,10))
T1=tuple(filter(odd,T))
print(T1)
```

```
#filter odd values from tuple of 10 range using lambda

T2=tuple(filter(lambda x:x%2 != 0,T))
print(T2)
```
